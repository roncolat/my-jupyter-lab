# Visualizzazione Scatter Plot Interattivo
Creare un grafico scatter utilizzando plotly.graph_objects con l'opzione 'lasso select' abilitata.

```python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

# Generazione dei dati casuali
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Variabile per memorizzare i punti selezionati
selected_points = []

# Creazione del grafico scatter interattivo
fig = go.FigureWidget(
    data=[
        go.Scatter(
            x=x,
            y=y,
            mode='markers',
            marker=dict(size=10, color='blue'),
            selected=dict(marker=dict(color='red')),
            unselected=dict(marker=dict(color='blue', opacity=0.3))
        )
    ],
    layout=go.Layout(
        title='Scatter Plot Interattivo con Selezione Lasso',
        dragmode='lasso'
    )
)

# Funzione per gestire la selezione dei punti
def on_selection(trace, points, selector):
    selected_indices = points.point_inds
    global selected_points
    selected_points = {'x': x[selected_indices], 'y': y[selected_indices]}
    print("Punti selezionati:", selected_points)

# Collegamento della funzione di selezione al grafico
fig.data[0].on_selection(on_selection)

fig
```
